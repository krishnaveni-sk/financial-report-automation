```python
import pandas as pd

data = {
    "Customer_ID": [1, 2, 3, 2, 4],
    "Customer_Name": ["Amit", "Ravi", "Sita", "Ravi", "John"],
    "Region": ["South", "North", "South", "North", "West"],
    "Amount": [5000, 7000, None, 7000, 6500],
    "Date": ["2024-01-02", "2024-01-03", "2024-01-05", "2024-01-03", "2024-01-07"]
}

df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)
df
```

```python
df = pd.read_csv("sales_data.csv")
df = df.drop_duplicates()
df['Amount'] = df['Amount'].fillna(0)
df['Date'] = pd.to_datetime(df['Date'])
df.to_csv("cleaned_sales_data.csv", index=False)
df
```

```python
summary = df.groupby('Region')['Amount'].sum().reset_index()
summary.to_csv("summary_report.csv", index=False)
summary
```
